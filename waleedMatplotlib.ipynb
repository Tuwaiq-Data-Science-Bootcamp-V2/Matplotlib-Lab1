## Q1: Create two charts using list and dataframe

```python
onelist = {'Cars':['BMW','Lkxz','Ferary'],
          'SaleAmount':[439,672,300],
          'Years':[2011,2012,2013]} 
test = pd.DataFrame(onelist,columns=onelist.keys())
plt.subplot(1,2,1)
plt.hist(test['SaleAmount'])
plt.title('Saleing Cars')
plt.subplot(1,2,2)
plt.plot(test['SaleAmount'])
plt.title('chart2')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(onelist['Years'], autopct='%1.1f%%',explode = [0,0,0])
plt.title('Pie chart')
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(test['Years'],test['SaleAmount'],s = 50, alpha = 0.4)
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.title('test',loc='left',c='red')
plt.xlabel('x label',fontsize = 16)
plt.ylabel('y label',fontsize = 16)
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot(test['Years'],color='Grey')
plt.legend(['Years'])
```

## Q6: Create subplots with three charts and figsize=20,10

```python

a1 , a2 =plt.subplots(3, sharex=True, sharey=True,figsize=(20,10))
a1.suptitle('testing')
a2[0].plot(test['Years'],color='Grey')
a2[1].hist(test['Years'])
a2[2].pie(onelist['Years'], autopct='%1.1f%%')
#plt.show()
```
