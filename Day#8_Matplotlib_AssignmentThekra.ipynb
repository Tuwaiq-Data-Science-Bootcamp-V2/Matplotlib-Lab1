```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
price = [2.50, 1.23, 4.02, 3.25, 5.00, 4.40]
sales_per_day = [34, 62, 49, 22, 13, 19]
profit_margin = np.asarray([20, 35, 40, 20, 27.5, 15])
plt.scatter(x=price, y=sales_per_day, alpha=0.4,s=profit_margin*50)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.figure(figsize=(10,6))
plt.plot(np.arange(1,8),c='b',label='A')
plt.plot(np.arange(8,11),c='g',label='B')
plt.plot(np.arange(4,7),c='y',label='C')
plt.legend(loc='upper left')
plt.show()
```
