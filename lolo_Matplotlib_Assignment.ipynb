```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
animals = ['duck','dog','cat']
numbers = [7,6,5]
plt.bar(animals,numbers)
plt.xlabel('The animals')
plt.ylabel('Numbers of Animals')
plt.title('Number of Animals in the pet shop')
plt.show()
```

```python
data = {
    "Name" : ['Sam','Joe','Alix'],
  "calories": [420, 380, 390]
}


df = pd.DataFrame(data)
plt.bar(df['Name'],df['calories'])
plt.xlabel('Names')
plt.ylabel('Numbers of calories')
plt.title('Number of calories in the persons')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
animals = ['duck','dog','cat']
numbers = [7,6,5]
colors = ['#ff9999','#66b3ff','#99ff99']
plt.pie(numbers,labels=animals, autopct='%1.1f%%', colors =colors,  startangle=180, radius=1.5)

plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
data = {
    "Name" : ['Sam','Joe','Alix'],
  "calories": [420, 380, 390]
}


df = pd.DataFrame(data)
plt.scatter(df['Name'],df['calories'],alpha = 0.4,s=100)
plt.xlabel('Names')
plt.ylabel('Numbers of calories')
plt.title('Number of calories in the persons')
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
data = {
    "Name" : ['Sam','Joe','Alix'],
  "calories": [420, 380, 390]
}


df = pd.DataFrame(data)
plt.scatter(df['Name'],df['calories'],alpha = 0.4,s=100)
plt.xlabel('Names',size = 16)
plt.ylabel('Numbers of calories', size =16)
plt.title('Number of calories in the persons', ha = "right", color = "red")
plt.show()
```

## Q5: Create three line charts with Legends at left and colors

```python
data = {
    "Name" : ['Sam','Joe','Alix'],
  "calories": [420, 380, 390],
    "duration": [50, 40, 45],
    "Age" : [16,15,12]
}


df = pd.DataFrame(data)
plt.plot(df['Name'],df['Age'])
plt.plot(df['Name'],df['calories'])
plt.plot(df['Name'],df['duration'])
plt.legend(['Age','calories','duration'], shadow = True, loc = 6) 
plt.show()
```
