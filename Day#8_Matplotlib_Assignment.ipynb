```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
data={'course':['data science','AI','MATH','Biosciece','STAT'],'avrage grade':[99.5,75.6,66.4,78.3,81.4],'student count':[25,40,55,32,15]}
df=pd.DataFrame(data,columns=data.keys())
df
```

```python
plt.bar(df['course'],df['avrage grade'],color=['r','b','g'])
plt.title('avrage grade for every course')
plt.ylabel('avrage')
plt.xlabel('course')
plt.show()
```

```python
plt.bar(data['course'],data['avrage grade'],color=['r','b','g'])
plt.title('avrage grade for every course')
plt.ylabel('avrage')
plt.xlabel('course')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
plt.figure(figsize=(10,10))
plt.pie(df['student count'],radius=1,autopct='%1.1f%%',explode=[0,0,0,0,0])
plt.legend(df['course'],loc=0)
plt.title('precentage of every course')
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.figure(figsize=(10,10))
plt.scatter(df['avrage grade'],df['student count'],alpha=0.4 , marker='o',s=50,color='r')
plt.title('correlation between student count and grades')
plt.ylabel('studen count')
plt.xlabel('average grade')
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.figure(figsize=(10,10))
plt.scatter(df['avrage grade'],df['student count'],alpha=0.4 , marker='o',s=50,color='r')
plt.title('correlation between student count and grades',loc='left')
plt.ylabel('studen count',fontdict={'size':16})
plt.xlabel('average grade',fontdict={'size':16})
plt.show()
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.figure(figsize=(5,5))
plt.plot(df['avrage grade'],df['student count'], marker='o',color='r')
plt.title('correlation between student count and grades',loc='left')
plt.ylabel('studen count',fontdict={'size':16})
plt.xlabel('average grade',fontdict={'size':16})
plt.show()

plt.figure(figsize=(5,5))
plt.plot(df['avrage grade'],df['student count'], marker='o',color='b')
plt.title('correlation between student count and grades',loc='left')
plt.ylabel('studen count',fontdict={'size':16})
plt.xlabel('average grade',fontdict={'size':16})
plt.show()

plt.figure(figsize=(5,5))
plt.plot(df['avrage grade'],df['student count'], marker='o',color='g')
plt.title('correlation between student count and grades',loc='left')
plt.ylabel('studen count',fontdict={'size':16})
plt.xlabel('average grade',fontdict={'size':16})
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])
plt.suptitle('line charts',fontsize = 20)

plt.subplot(1,3,1)
plt.plot(df['avrage grade'],df['student count'], marker='o',color='r')
plt.title('correlation between student count and grades',loc='left')
plt.ylabel('studen count',fontdict={'size':16})
plt.xlabel('average grade',fontdict={'size':16})

plt.subplot(1,3,2)
plt.plot(df['avrage grade'],df['student count'], marker='o',color='b')
plt.title('correlation between student count and grades',loc='left')
plt.ylabel('studen count',fontdict={'size':16})
plt.xlabel('average grade',fontdict={'size':16})

plt.subplot(1,3,3)
plt.plot(df['avrage grade'],df['student count'], marker='o',color='g')
plt.title('correlation between student count and grades',loc='left')
plt.ylabel('studen count',fontdict={'size':16})
plt.xlabel('average grade',fontdict={'size':16})
```
