```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
movies_names = ['avatar','star wars','paris, texas', 'old boy']
movies_rating = [7.2,8.1,10,9.4]

plt.bar(movies_names,movies_rating)
plt.title('Rating of movies')
plt.xlabel('movies names')
plt.ylabel('movies rating')

plt.show()
```

```python
movies_data = {'name' : ['avatar','star wars','paris, texas', 'old boy'],
                  'rating' : [7.2,8.1,10,9.4]
                 }
movies = pd.DataFrame(data=movies_data, columns=['name','rating'])
plt.bar(movies['name'],movies['rating'])
plt.title('Rating of movies')
plt.xlabel('movies names')
plt.ylabel('movies rating')

plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
name = ['elon musk','jeff bezos','mohammed alrajeh','bill gates']
wealth = [ 20323889661,19582457291,115245724341,39287624062]

plt.title('wealth distribution')
plt.pie(wealth,labels=name, autopct='%1.1f%%')

plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
height=[160,166,170,182,169,174,156,190,145,166]
weight=[70,57,66,69,89,77,54,80,77,68]
age=[20,50,27,33,32,40,56,19,34,29]
plt.scatter(height,weight,alpha=0.4)
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
height=[160,166,170,182,169,174,156,190,145,166]
weight=[70,57,66,69,89,77,54,80,77,68]
plt.scatter(height,weight,alpha=0.4)
plt.title('the relationship between height and wight',loc='left',c='r')
plt.xlabel('height',fontsize=16)
plt.ylabel('weight',fontsize=16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot(height,weight, c = '0.4')
plt.plot(weight,height, c = '0.4')
plt.plot(age,weight, c = '0.4')
plt.legend(['height','weight','age'],loc = 'upper left')
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])
plt.suptitle('three charts',fontsize = 20)

#first
plt.subplot(1,3,1) #(number of rows, number of columns, number of plot)
plt.title('wealth distribution')
plt.pie(wealth,labels=name, autopct='%1.1f%%')

#second
plt.subplot(1,3,2)
plt.scatter(height,weight,alpha=0.4)

#third
plt.subplot(1,3,3)
plt.plot(height,weight, c = '0.4')
plt.plot(weight,height, c = '0.4')
plt.plot(age,weight, c = '0.4')
plt.legend(['height','weight','age'],loc = 'upper left')

plt.show()
```
