## Q2: Create pie chart with equal explode values.

```python
df.head()
df.groupby('real_fake').sum().sort_values(by='Post interval',ascending = False)['Post interval']
realfake = ['real','fake']
postint = [2.092423e+07,1.150851e+07]
colors =['y','g']
plt.pie(postint,labels=realfake, autopct='%1.2f%%', colors =colors,  startangle=180, radius=1.5,explode=[0.1,0.1])
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
dfreal = df[df['real_fake']=='real']
plt.scatter(dfreal['Num_following'],dfreal['Post interval'],alpha = 0.3)
plt.xlabel('Num_following',fontsize = 20)
plt.ylabel('Post interval in Hour',fontsize = 20)
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(dfreal['Num_following'],dfreal['Post interval'],alpha = 0.3)
plt.xlabel('Num_following',fontsize = 16)
plt.ylabel('Post interval in Hour',fontsize = 16)
plt.title('Numbers following in average post interval for real accounts',c='r',loc='left')
plt.show
```

## Q5: Create three line charts with Legends at left and colors

```python
x1 ,y1= [1,2,3,4,5,6] , [1,6,8,15,20,30]
y2 = [3,6,7,12,15,23]
y3 = [12,9,8,6,5,1]
plt.plot(x1,y1,color='red')
plt.plot(x1,y2,color = 'Blue')
plt.plot(x1,y3, 'Green')
plt.legend(['Fire Red','Frozen Blue','Tree Green'], loc='upper left')
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10

```python

fig ,ax = plt.subplots(3,1,sharey=True,figsize=(20,10))
x1 ,y1= [1,2,3,4,5,6] , [1,6,8,15,20,30]
y2 = [3,6,7,12,15,23]
y3 = [12,9,8,6,5,1]
ax[0].plot(x1,y1,color='red')
ax[1].plot(x1,y2,color = 'Blue')
ax[2].plot(x1,y3, 'Green')
plt.show()
```
