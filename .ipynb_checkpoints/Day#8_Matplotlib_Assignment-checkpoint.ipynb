```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
df ={'StudentName':['manar','rahaf','khalda','lamess'],
    'year':[1999,2000,1997,1998],
    'age':[22,23,25,26]}

plt.bar(df['StudentName'], df['age'])
plt.xlabel('Age',fontsize = 20)
plt.ylabel('Name',fontsize = 20)
plt.show()
plt.plot(df['StudentName'], df['age'])

plt.xlabel('Age',fontsize = 20)
plt.ylabel('Name',fontsize = 20)







plt.show()


```

## Q2: Create pie chart with equal explode values.

```python
sizes = [25, 25, 25, 25]

labels = ['team1','team2','team3','team4']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python


plt.scatter(df['year'],df['age'], s=50 ,alpha = 0.4) 


plt.xlabel('age',fontsize = 20)
plt.ylabel('year',fontsize = 20)


plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python


plt.scatter(df['year'],df['age'], s=50 ,alpha = 0.4) 
plt.title('student', color = 'red' ,loc='left')
plt.xlabel('age',fontsize = 16)
plt.ylabel('year',fontsize = 16)


plt.show()
```

## Q5: Create three line charts with Legends at left and colors

```python

plt.plot(df['year'],df['age'], c ='red')

plt.legend(['age'],  loc = 2) 


plt.show()

rev = [1000,1500,2000]
yeas = [1990, 1995, 2000]


plt.plot(yeas,rev, c ='green')

plt.legend(['revenue'], loc = 2) 


plt.show()

grade=[5,6,7]
clas = [1,2,3]

plt.plot(clas,grade, c ='blue')

plt.legend(['garde'],  loc = 2) 


plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=(20,10))

#first plot
plt.subplot(3,1,1) 
plt.bar(df['StudentName'], df['age'])
plt.xlabel('Age',fontsize = 20)
plt.ylabel('Name',fontsize = 20)
plt.show()


plt.subplot(3,1,2) 

plt.plot(df['year'],df['age'], c ='red')

plt.xlabel('Year',fontsize = 20)
plt.ylabel('Age',fontsize = 20)
plt.show()





plt.subplot(3,1,3) 


plt.scatter(df['year'],df['age'], s=50 ,alpha = 0.4) 
plt.title('student', color = 'red' ,loc='left')
plt.xlabel('age',fontsize = 16)
plt.ylabel('year',fontsize = 16)


plt.show()
```
